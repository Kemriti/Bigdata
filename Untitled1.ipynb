```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
plt.style.use("seaborn")
```

```python
data = pd.read_csv("immobilier_france.csv")
```

```python
Rennes = data[data["ville"]=="RENNES"]
```

```python
Rennes.index=Rennes["date_transaction"]
```

```python
Rennes['date_transaction'] = pd.to_datetime(Rennes['date_transaction'])

Rennes['annee_mois'] = Rennes['date_transaction'].dt.strftime('%Y-%m')
volume = Rennes.groupby('annee_mois').size().reset_index(name='Volume de transaction/mois')
Rennes = pd.merge(Rennes, volume, on='annee_mois', how='right')



```

```python
resultat = Rennes.groupby('annee_mois')['prix_m2'].mean().reset_index()
Rennes = pd.merge(Rennes, resultat, on='annee_mois', how='right')
Rennes = Rennes.rename(columns={'prix_m2_y': 'moy prix m²/mois'})
```

```python
Rennes = Rennes.drop(["surface_dependances","surface_locaux_industriels","surface_terrains_agricoles","surface_terrains_sols","surface_terrains_nature"],axis=1)
```

```python
import requests
import pandas as pd
url = 'https://www.insee.fr/fr/statistiques/7662181#consulter'

# Envoyer une requête GET pour récupérer la page web
response = requests.get(url)

# Vérifier si la requête a réussi
if response.status_code == 200:
    # Lire le tableau des données en utilisant pandas
    dfs = pd.read_html(response.text)
    # Supposons que le premier tableau contient les données d'inflation
    inflation_data = dfs[1]
    
    # Vous pouvez maintenant travailler avec les données d'inflation
    print(inflation_data)
else:
    print("Échec de la requête HTTP")
```

```python
inflation_data.index=inflation_data["Unnamed: 0"]
```

```python
inflation_data["Glissement annuel"]
```

```python
import pandas as pd

# Supposons que votre DataFrame est appelé "inflation_data"
# Assurez-vous que la colonne "Date" est au format datetime
inflation_data["Date"] = pd.to_datetime(inflation_data["Unnamed: 0"])

# Créez une condition pour filtrer les données de janvier 2014 jusqu'à décembre 2022
start_date = pd.to_datetime("2014-01-01")
end_date = pd.to_datetime("2022-12-31")

# Appliquez la condition pour extraire les données dans la plage spécifiée
filtered_data = inflation_data[(inflation_data["Date"] >= start_date) & (inflation_data["Date"] <= end_date)]

# Maintenant, filtered_data contient uniquement les données de janvier 2014 jusqu'à décembre 2022
# dans la colonne "Glissement annuel"
```
